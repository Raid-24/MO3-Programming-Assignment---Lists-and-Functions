9.2: Define a generator function called get_odds() that returns the odd numbers from range(10). Use a for loop to find and print the third value returned.

```python
def get_odds():
    for number in range(1,10):
        if number % 2 != 0:
            yield number

odd_nums = get_odds()
step = 0

for odd_num in odd_nums:
    step += 1
    if step == 3:
        third_value = odd_num
        print(odd_num)
        break
```
